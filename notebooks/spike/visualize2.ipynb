```python
def get_sex_Xtarget(data):
    return visualize.get_Xtarget(data, 'sex_male', ['credit', 'sex_female'])

def get_age_cat_Xtarget(data):
    return visualize.get_Xtarget(data, 'age_cat_old', ['credit', 'age', 'age_cat_young'])

def compare_by(data, repaired_data, get_Xtarget=get_sex_Xtarget):
    X, target = get_Xtarget(data)
    goal = data['credit']
    xy = visualize.project2d_classifier_goal(X, target, goal)
    #xy = visualize.project2d_classifier(X, target)
    #xy = visualize.project2d(X)
    visualize.plot2(xy, target, goal)
    
    X, target = get_Xtarget(repaired_data)
    xy = visualize.project2d_classifier_goal(X, target, goal)
    #xy = visualize.project2d_classifier(X, target)
    #xy = visualize.project2d(X)
    visualize.plot2(xy, target, goal)
    
```

```python
import pandas as pd
import BlackBoxAuditing as bba
```

```python
n = 500
p = 0.5
age_group = np.random.binomial(n=1, p=p, size=n)
age = np.random.normal(25, 5, size=n) * age_group + np.random.normal(65, 10, size=n) * (1 - age_group)
knowledge = np.maximum(np.random.normal(10, 5, size=n) + age/10, 2)
time_in_job = np.random.exponential(2, size=n) * age_group + \
    np.random.exponential(5, size=n) * (1 - age_group)
income = np.maximum((knowledge + time_in_job / 3) * np.maximum(np.random.normal(100, 50), 100), 1000)
credit = np.logical_and(income >= 1500, time_in_job >= 0.5)
age_cat = age >= 30

synthetic = pd.DataFrame()
synthetic['age_cat'] = 1 * age_cat
synthetic['knowledge'] = knowledge
synthetic['time_in_job'] = time_in_job
synthetic['income'] = income
synthetic['credit'] = 1 * credit
```

```python
synthetic
```

```python
repairer = bba.repairers.GeneralRepairer.Repairer(synthetic.values.tolist(), 0, 0.6, True, features_to_ignore = [4])
repaired_synthetic = pd.DataFrame.from_records(repairer.repair(synthetic.values.tolist()), columns=['age_cat', 'knowledge', 'income', 'time_in_job', 'credit'])
repaired_synthetic['age_cat'] = synthetic['age_cat']
```

```python
compare_by(synthetic, repaired_synthetic, get_Xtarget=lambda data: visualize.get_Xtarget(data, 'age_cat', ['credit']))
```

```python
plt.plot(age, 'x')
plt.show()
plt.plot(income, 'x')
plt.show()
plt.plot(time_in_job, 'x')
plt.show()
plt.plot(credit, 'x')
plt.plot(age_cat, 'ro')
print(np.count_nonzero(credit))
print(np.count_nonzero(age_cat))
```
